```python
 # Libraries
import pandas as pd
```

```python
# importing dataset and assigning them into a new variable name users
URL = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(URL, sep ="|")
users.head(10)
```

```python
# Discover what is the mean age per occupation
users.groupby('occupation').age.mean()
```

```python
#Discover the Male ratio per occupation and sort it from the most to the least
def gender(var):
    if var == 'M' :
        return True
    else :
        return False
    
users['maleOnly'] = users['gender'].apply(gender)
male_number = users.groupby('occupation')['maleOnly'].sum()
male_number
```

```python
# Taking total
total_number = users.groupby('occupation').gender.count()
total_number
```

```python
#sorting from high to low
ratio = male_number/total_number
ratioSort = ratio.sort_values(ascending = False)
ratioSort
```

```python
# For each occupation, calculate the minimum and maximum ages
users.groupby('occupation').age.min()
```

```python
users.groupby('occupation').age.max()
```

```python
#For each combination of occupation and sex, calculate the mean age
users.groupby(['occupation','gender'])['age'].mean()
```

```python
#For each occupation present the percentage of women and men
people_gender = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
people_gender
```

```python
# Number of employee in an occupation
occupation = users.groupby(['occupation']).count()
occupation
```

```python
Percentage_of_each_gender = people_gender.div(occupation, level = "occupation")*100
Percentage_of_each_gender.loc[:, 'gender']
```
